# Importing Libraries

```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, accuracy_score
import pickle
```

```python
PATH = '/content/drive/My Drive/predicting graduate admission/'
```

```python
df = pd.read_csv(PATH + 'input/train.csv')
```

```python
df.head()
```

```python
df.drop('Serial No.', axis=1, inplace=True)
df.head()
```

# Training and Test Set

```python
X = df.drop('Chance of Admit ', axis=1)
y = df['Chance of Admit ']
```

```python
X = np.array(X)
y = np.array(y)
```

```python
y = y.reshape(-1,1)
```

```python
# scaling the data before training the model
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler_x = StandardScaler()
X = scaler_x.fit_transform(X)
```

```python
scaler_y = StandardScaler()
y = scaler_y.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)
```

# Linear Regression

```python
linear_regression_model = LinearRegression()
linear_regression_model.fit(X_train, y_train)
```

```python
accuracy_LinearRegression = linear_regression_model.score(X_test, y_test)
accuracy_LinearRegression
```

# Lasso Regression

```python
# create a lasso regressor
lasso = Lasso(alpha=0.2)

# Fit the regressor to the data
lasso.fit(X_train,y_train)

# Compute and print the coefficients
lasso_coef = lasso.coef_
print(lasso_coef)
```

```python
accuracy_lasso = lasso.score(X_test, y_test)
accuracy_lasso
```

```python
pickle.dump(lasso, open('/content/drive/MyDrive/predicting graduate admission/saved_models/lasso.pkl', 'wb'))
```

# Ridge Regression

```python
ridge = Ridge(alpha=0.2)

ridge.fit(X_train, y_train)

ridge_coef = ridge.coef_
print(ridge_coef)
```

```python
accuracy_ridge = ridge.score(X_test, y_test)
accuracy_ridge
```
